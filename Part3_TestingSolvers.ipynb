# Dijkstra

```python
def dijkstra(source,graph):

    distance_dict={}
    min_dist_heap=[]

    for node, nbors in graph.items():
        distance_dict[node]=(float('inf'),-1) #(min_dist,parent)

    #distance_dict[source]=(0,source)  #<current_min_dist, parent>
    distance_dict[source]=(0,-1)
    heapq.heappush(min_dist_heap,(distance_dict[source][0],source))  #HeapEntry: <distance,vertex>


    while(min_dist_heap):  #while heap not empty
        curr_min,curr_vertex=heapq.heappop(min_dist_heap) # Pop minimum value entry
        if(curr_min>distance_dict[curr_vertex][0]): #check if it qualifies for propagation
            continue

        #If Qualified for propagation
        for nb, wt in graph[curr_vertex].items(): #update neigbours
            if(curr_min+wt<distance_dict[nb][0]): # if we get a strictly lesser value for a neigbour
                  distance_dict[nb]=(curr_min+wt,curr_vertex)  #update the neighbours distance dict
                  heapq.heappush(min_dist_heap,(curr_min+wt,nb))  #Add neighbour entry to heap
    return distance_dict

def track_SP(source,target,distances):
  node=target
  node_track=[]
  edge_track=[]
  while(node!=source):
    node_track.append((node,distances[node][0]))
    edge_track.append((distances[node][1],node))
    node=distances[node][1]
  node_track.append((node,distances[node][0]))
  node_track.reverse()
  edge_track.reverse()
  return node_track,edge_track
```

```python
def dijkstra_allsp(source,graph):

    distance_dict={}
    min_dist_heap=[]

    for node, nbors in graph.items():
        distance_dict[node]=(float('inf'),[-1]) #(min_dist,parent)

    distance_dict[source]=(0,[-1])  #<current_min_dist, parent>
    heapq.heappush(min_dist_heap,(distance_dict[source][0],source))  #HeapEntry: <distance,vertex>


    while(min_dist_heap):  #while heap not empty
        curr_min,curr_vertex=heapq.heappop(min_dist_heap) # Pop minimum value entry  --node x is reachable at distance ditsx (distx,node)
        if(curr_min>distance_dict[curr_vertex][0]): #check if it qualifies for propagation
            continue

        #If Qualified for propagation
        for nb, wt in graph[curr_vertex].items(): #update neigbours
            if(curr_min+wt<distance_dict[nb][0]): # if we get a strictly lesser value for a neigbour
                  distance_dict[nb]=(curr_min+wt,[curr_vertex])  #update the neighbours distance dict
                  heapq.heappush(min_dist_heap,(curr_min+wt,nb))  #Add neighbour entry to heap
            elif(curr_min+wt==distance_dict[nb][0]): #if an equally good path comes up
                 distance_dict[nb][1].append(curr_vertex)


    return distance_dict
```

```python
def get_parent_dict_from_source(distances):
    parent_dict={}
    for k,v in distances.items():
        parent_dict[k]=v[1]
    return parent_dict


def print_all_paths_from_source(current_node,path_list,all_sp,parent_dict,source,target): #works for the source for which we ran dijkstra

       if(current_node==source):
          path_list.append(source)
          all_sp.append(copy.deepcopy(path_list))
          return

       path_list.append(current_node)
       parents=parent_dict[current_node]
       for p in parents:
           current_node=p
           print_all_paths_from_source(current_node,path_list,all_sp,parent_dict,source,target)
           path_list.pop()
```

# Weighted BFS

```python
def weighted_bfs_orig_v1(graph,source):

    parent={}
    distance={}
    for node, edges in graph.items():
        parent[node]=-1
        distance[node]=-1


    Q=deque()
    distance[source]=0
    Q.append(source)

    while Q:
      node=Q.popleft()
      for nb, wt in graph[node].items():
          if(distance[nb]==-1):  #new
            distance[nb]=distance[node]+wt
            parent[nb]=node
            Q.append(nb)


    return  distance,parent
```

```python
def print_path(dest,parent):
    prev=dest
    path=[]
    while prev!=-1:
        path.append(str(prev))
        prev=parent[prev]

    path.reverse()
    print(" : ".join(path))

def find_path(dest,parent):
    prev=dest
    path=[]
    while prev!=-1:
        path.append(prev)
        prev=parent[prev]

    return path

def weighted_bfs_orig(graph,source):

    parent={}
    distance={}
    for node, edges in graph.items():
        parent[node]=-1
        distance[node]=float('inf')


    Q=deque()
    distance[source]=0
    parent[source]=-1
    Q.append(source)

    while Q:
      node=Q.popleft()
      for nb, wt in graph[node].items():
          if(distance[nb]==float('inf')):  #new
            distance[nb]=distance[node]+wt
            parent[nb]=node
            Q.append(nb)


    return  distance,parent
```

```python
def weighted_bfs_randshuffle(graph,source):

    parent={}
    distance={}
    for node, edges in graph.items():
        parent[node]=-1
        distance[node]=float('inf')


    Q=deque()
    distance[source]=0
    parent[source]=-1
    Q.append(source)

    while Q:
      node=Q.popleft()
      children=[]
      for nb, wt in graph[node].items():
            children.append((nb,wt))
      random.shuffle(children)
      for (child,wt) in children:
          if(distance[child]==float('inf')):  #new
              distance[child]=distance[node]+wt
              parent[child]=node
              Q.append(child)


    return  distance,parent
```

```python
def read_json(file_name):
    with open(file_name, 'r') as file:
        data = json.load(file)
    return data
```

```python
def convert_graph_to_int(graph_ob):
    data = {int(node): {int(k): int(v) for k, v in nbors.items()} for node, nbors in graph_ob.items()}
    return data

def convert_list_to_int(list_ob):
    data=[int(x) for x in list_ob]
    return data

def convert_listoflist_to_int(list_ob):
    data=[convert_list_to_int(x) for x in list_ob]
    return data
```

```python
def dfs(graph,source,visited,reachable):
    visited[source]=1
    reachable.append(source)
    for nbor in graph[source]:
        if(visited[nbor]==0):
           visited[nbor]=1
           dfs(graph,nbor,visited,reachable)

def find_reachable_vertices(source,vertex_list,graph):
    visited={i:0 for i in vertex_list} #visited array
    reachable=[] #nodes reachable from a source
    dfs(graph,source,visited,reachable)
    return reachable
```

```python
def verify_subpath_in_dij(dij_sp,test_graph):
    print("Verifying: Solver 1->Dijkstra")
    dij_spr=[x[::-1] for x in dij_sp]
    #Chose a random subpath
    # Chose a list
    r1=random.randint(0, len(dij_spr)-1)
    chosen_path=dij_spr[r1]
    print(f"Dijkstra chosen path")
    print(chosen_path)
    len_list=len(chosen_path)
    last_index=len_list-1
    #chose sub path in a list
    source_index=random.randint(0,len_list-1)
    pt_target_index=source_index+random.randint(0,len_list-1)
    target_index=last_index if pt_target_index>last_index else pt_target_index
    source,target=chosen_path[source_index],chosen_path[target_index]
    chosen_sub_path=chosen_path[source_index:target_index+1]
    print("Dijkstra chosen subpath to test")
    print(chosen_sub_path)
    #Run dijkstra for new source and target
    distances=dijkstra_allsp(source,test_graph)
    all_sp,path_list=[],[]
    parent_dict=get_parent_dict_from_source(distances)
    print_all_paths_from_source(target,path_list,all_sp,parent_dict,source,target)
    all_sp=[x[::-1] for x in all_sp]
    print(f"All paths from {source} to {target}")
    print(all_sp)
    if chosen_sub_path in all_sp:
       return True
    return False
```

```python
def verify_subpath_in_bfs(bfs_sp,test_graph):
    print("Verifying: Solver 2->WBFS")
    bfs_spr=bfs_sp[::-1]
    chosen_path=bfs_spr
    print(f"WBFS chosen path")
    print(chosen_path)
    len_list=len(chosen_path)
    last_index=len_list-1
    #chose sub path in a list
    source_index=random.randint(0,len_list-1)
    pt_target_index=source_index+random.randint(0,len_list-1)
    target_index=last_index if pt_target_index>last_index else pt_target_index
    source,target=chosen_path[source_index],chosen_path[target_index]
    chosen_sub_path=chosen_path[source_index:target_index+1]
    print("WBFS chosen subpath to test")
    print(chosen_sub_path)
    #Run bfs for new source and target

    dist_bfs,parents_bfs=weighted_bfs_randshuffle(test_graph,source) #------> randshuffle bfs
    #dist_bfs,parents_bfs=weighted_bfs_orig(test_graph,source)
    bfs_path=find_path(target,parents_bfs)
    bfs_pathr=bfs_path[::-1]
    print(f"WBFS path from {source} to {target}")
    print(bfs_pathr)
    if(bfs_pathr==chosen_sub_path):
      return True
    return False
```

```python
def compare_dij_op(dij1,dij2,rv_graph1,rv_graph2):
    newly_reachable=list(set(rv_graph2) - set(rv_graph1))
    flag_prev=True
    flag_new=True
    diff=0
    for  node in rv_graph1:
         if(dij1[node][0]!=dij2[node][0]): #if shortest distnaces dont match
            flag_prev=False
            diff=diff+1

    for  node in newly_reachable:
         if(dij2[node][0]>dij1[node][0]): #if shortest distnaces dont match
            flag_new=False
            diff=diff+1

    ret_flag=flag_prev and flag_new

    return ret_flag, diff
```

```python
def compare_wbfs_op(wbfs_g1,wbfs_g2,rv_graph1,rv_graph2):
    newly_reachable=list(set(rv_graph2) - set(rv_graph1))
    flag_prev=True
    flag_new=True
    diff=0
    for  node in rv_graph1:
         if(wbfs_g1[node]!=wbfs_g2[node]): #if shortest distnaces dont match
            flag_prev=False
            diff=diff+1

    for  node in newly_reachable:
         if(wbfs_g2[node]>wbfs_g1[node]): #if shortest distnaces dont match
            flag_new=False
            diff=diff+1

    ret_flag=flag_prev and flag_new

    return ret_flag, diff
```

```python
def check_all_distances(dij_dist,wb_dist,source_list):
    #dij_dist entry: Vertex: (Min_dist_source,parent)
    #wb_dist entry: wb_dist[node]=min_dist_source
    vertex_list=source_list
    diff=0
    flag=True
    for  node in vertex_list:
         if(dij_dist[node][0]!=wb_dist[node]): #if shortest distnaces dont match
            flag=False
            diff=diff+1

    return flag,diff,len(vertex_list)
```

# Test Case 4:
$SSSP(a,x) + SSSP(x,b) \geq SSSP(a,b)$

This test case is passing always for Dijkstra and sometimes failing for WBFS, this is because when calculating the distance from a to a, the # of nodes traversed is considered -1, rather than 0


```python
def addIntermediate(test_data):
  dijk = False
  bfs = False
  print("Original Graph:")
  print(test_data)

  # choose a random a, b, x
  a = 0
  # a = random.randint(0, len(test_data) - 1)
  b = random.randint(0, len(test_data) - 1)
  x = random.randint(0, len(test_data) - 1)

  print("\nNodes:")
  print('A:', a, 'B:', b, 'X:', x, '\n')

  # first run through dijkstra's: find SSSP(a, b) and SSSP(a, x)
  distance_total = dijkstra_v1(a, test_data)
  #print(distance_total)
  distanceAB = distance_total[b][0]
  distanceAX = distance_total[x][0]

  # second run through dijkstra's : find SSSP(x,b)
  distance_secondary = dijkstra_v1(x, test_data)
  distanceXB = distance_secondary[b][0]

  if(distanceAX + distanceXB >= distanceAB):
    print("Test Case PASSED - Dijkstra")
  else:
    dijk = True
    print("Test Case FAILED - Dijkstra")

  # now do the same thing with wbfs:
  # first run through: find SSSP(a, b) and SSSP(a, x)
  distance_total = weighted_bfs_orig_v1(test_data, a)
  distanceAB = distance_total[1][b]
  distanceAX = distance_total[1][x]

  #second run through: find SSSP(x,b)
  distance_secondary = weighted_bfs_orig_v1(test_data, x)
  distanceXB = distance_secondary[1][b]

  if(distanceAX + distanceXB >= distanceAB):
    print("Test Case PASSED - Weighted BFS")
  else:
    bfs = True
    print("Test Case FAILED - Weighted BFS")

  if (dijk and bfs) or (not dijk and not bfs):
    print("The two implementations are the same")
  else:
    print("The two implementations are different")
  #eof

addIntermediate(test_data['graph'])
```
